```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
```

```python
img1=cv2.imread("/content/sample_data/P1.jpg")
img2=cv2.imread("/content/sample_data/P2.jpg")
cv2_imshow(img1)
cv2_imshow(img2)
```

Addition images


```python
img1=cv2.resize(img1,(300,300))
img2=cv2.resize(img2,(300,300))
img_add=cv2.add(img1,img2)
cv2_imshow(img_add)
```

```python
img_add1=cv2.addWeighted(img1,0.5,img2,0.2,0)
cv2_imshow(img_add1)
```

Subtraction

```python
img_sub=cv2.subtract(img1,img2)
cv2_imshow(img_sub)
```

Multiplication

```python
img_mul=cv2.multiply(img1,img2)
cv2_imshow(img_mul)
```
